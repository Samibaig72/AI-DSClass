```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= [2,4,3,4,6,5,3,5,6]
y= [1,3,5,6,6,7,7,8,9]

plt.scatter(x,y, label="data1", color='r', s=100, marker='.')
plt.plot(x,y)

plt.title("Saylani")
plt.legend()

for i,j in zip(x,y):
    plt.annotate((i,j),(i,j))

plt.show()
```
